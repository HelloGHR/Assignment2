**Second Approach for Speech to text**

```python
!pip install SpeechRecognition
import speech_recognition as sr
def SpeechTranslation(audio_file):
  import speech_recognition as sr
# Create a Recognizer instance
  r = sr.Recognizer()
# Load the audio file
  audio_file_path = audio_file
  audio_file = sr.AudioFile(audio_file_path)
  # Recognize speech from the audio file
  with audio_file as source:
    audio = r.record(source)
# Recognize speech using Google Speech Recognition
  try:
    # Recognize the speech with the language set to Hindi
    text = r.recognize_google(audio, language="hi-IN")
    print(f"Recognized text: {text}")
  except sr.UnknownValueError:
    print("Google Speech Recognition could not understand audio")
  except sr.RequestError as e:
    print(f"Could not request results from Google Speech Recognition service; {e}")




audio_file="/content/har_ni_ka_daar.wav"
SpeechTranslation(audio_file)
```

**First Approach for Text-To-Speech**

```python
!pip install pydub
!pip install git+https://github.com/suno-ai/bark.git
from bark import SAMPLE_RATE, generate_audio, preload_models
from pydub import AudioSegment
from IPython.display import Audio, display
import numpy as np

preload_models()

def TTS(text_prompt, max_length=250):
    def split_text(text, max_length=max_length):
        words = text.split()
        chunks = []
        current_chunk = ""
        for word in words:
            if len(current_chunk) + len(word) + 1 <= max_length:
                current_chunk += word + " "
            else:
                chunks.append(current_chunk.strip())
                current_chunk = word + " "
        if current_chunk:
            chunks.append(current_chunk.strip())
        return chunks

    text_chunks = split_text(text_prompt)

    combined_audio = AudioSegment.silent(duration=0)

    for chunk in text_chunks:
        if chunk.strip():  # Ensure the chunk is not empty or whitespace
            audio_array = generate_audio(chunk)
            audio_segment = AudioSegment(
                np.int16(audio_array * 32767).tobytes(),
                frame_rate=SAMPLE_RATE,
                sample_width=2,
                channels=1
            )
            combined_audio += audio_segment

    output_file = "final_output.wav"
    combined_audio.export(output_file, format="wav")
    display(Audio(output_file, rate=SAMPLE_RATE))

text_prompt = """मेरे प्यारे दोस्त, आप सभी कैसे हैं? मैं आपको इस प्यारे दिन में एक छोटा सा संदेश भेज रहा हूँ। आपका स्नेह और समर्थन मेरे लिए अत्यंत महत्वपूर्ण है। आप सबका धन्यवाद।
"""
TTS(text_prompt)
```

```python
!pip install gtts
!pip install playsound
from gtts import gTTS
from playsound import playsound
from IPython.display import Audio, display


def hindi_text_to_speech(hindi_text, output_file="hindi_speech.mp3"):
    tts = gTTS(text=hindi_text, lang='hi')
    tts.save(output_file)
    print(f"Hindi speech saved to {output_file}")
    display(Audio(output_file))



hindi_text = "नमस्ते, आप कैसे हैं?"
hindi_text_to_speech(hindi_text)
```
